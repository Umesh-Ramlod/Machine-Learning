```python
#Assignment no : 02
import pandas as pd
import numpy as np
from scipy.spatial import distance
```

```python
comedy_scenes = [100, 0, 15, 85, 10, 85, 6, 93, 50]
Action_scenes = [0, 100, 90, 20, 95, 15, 70, 23, 50]
Actual_class = ["Comedy", "Action", "Action", "Comedy", "Action", "Comedy", "", "", ""]
```

```python
train_data = pd.DataFrame({"Comedy_Scenes":comedy_scenes[0:4], "Action_Scenes":Action_scenes[0:4], "Actual_Class":Actual_class[0:4] })
train_data
```

```python
validation_data = pd.DataFrame({"Comedy_Scenes":comedy_scenes[4:6], "Action_Scenes":Action_scenes[4:6], "Actual_Class":Actual_class[4:6]})
validation_data
```

```python
test_data = pd.DataFrame({"Comedy_Scenes":comedy_scenes[6:], "Action_Scenes":Action_scenes[6:], "Actual_Class":Actual_class[6:]})
test_data
```

```python
def Euclidean_distance(a,b,x,y):
    d = ((a-x)**2 + (b-y)**2 ) ** 1/2
    return d
```

```python
distance = []
for i in range(2):
    a = validation_data.iloc[i]["Comedy_Scenes"]
    b = validation_data.iloc[i]["Action_Scenes"]
    result = []
    for j in range(4):
        output = {}
        x = train_data.iloc[j]["Comedy_Scenes"]
        y =  train_data.iloc[j]["Action_Scenes"]
        temp = Euclidean_distance(a,b,x,y)
        a_class = train_data.iloc[j]["Actual_Class"]
        output['value']=temp
        output['name']=a_class
        result.append(output)
    distance.append(result)
print(distance,sep="\n")
```

```python
# determine the value of k
accuracy = 0
for i in range(2):
    distance[i].sort(key=lambda k: k.get('value'))
for i in range(2):
    target = validation_data.iloc[i]["Actual_Class"]
    value_of_k =0
    for j in range(1,5):
        k = j
        match , a = 0, 0
        wrong_match = 0
        for m in range(k):
            values = distance[i][m]['name']
            if values == target:
                match += 1
            else:
                wrong_match += 1
        if(match>wrong_match):
             value_of_k =max(value_of_k,k)
print(value_of_k)                        
```

Here We have calculated the most favoraible value of k for our model...

```python
# now predict the outputu for test data

predict_distance = []
for i in range(3):
    a = test_data.iloc[i]["Comedy_Scenes"]
    b = test_data.iloc[i]["Action_Scenes"]
    result = []
    for j in range(4):
        output = {}
        x = train_data.iloc[j]["Comedy_Scenes"]
        y =  train_data.iloc[j]["Action_Scenes"]
        temp = Euclidean_distance(a,b,x,y)
        a_class = train_data.iloc[j]["Actual_Class"]
        output['value']=temp
        output['name']=a_class
        result.append(output)
    predict_distance.append(result)
for i in range(3):
    predict_distance[i].sort(key=lambda k: k.get('value'))
```

```python
for i in range(3):
    action , comedy = 0, 0
    predicted_output = ""
    for k in range(3):
        values = predict_distance[i][k]['name']
        if values == 'Action':
            action+=1
        else:
            comedy+=1
    if(action>=comedy):
        predicted_output = "Action"
    else:
        predicted_output = "Comedy"
    test_data.at[i,'Actual_Class']= predicted_output
```

# IRIS DATASET


```python
# Load the iris dataset into a dataframe

Iris_dataframe = pd.read_csv("./iris.csv")
Iris_dataframe.columns = ['Sepal_Length','Sepal_Width','Petal_Length','Petal_Width','Class']
for i in range(149):
    Iris_dataframe.at[i,"Class"] = Iris_dataframe.iloc[i]["Class"].replace("Iris-","")
```

```python
# Split the dataset into 60% 20% 20% ratio...

   
train, validate , test = np.split(Iris_dataframe.sample(frac=1,random_state=24),[int(.6*len(Iris_dataframe)),int(.8*len(Iris_dataframe))])
```

```python
distance_iris = []
for i in range(len(validate)):
    a = validate.iloc[i]["Sepal_Length"]
    b= validate.iloc[i]["Sepal_Width"]
    c = validate.iloc[i]["Petal_Length"]
    d = validate.iloc[i]["Petal_Width"]
    val_array = np.array((a,b,c,d))
    result = []
    for j in range(len(train)):
        output = {}
        w = train.iloc[j]["Sepal_Length"]
        x = train.iloc[j]["Sepal_Width"]
        y = train.iloc[j]["Petal_Length"]
        z = train.iloc[j]["Petal_Width"]
        train_array = np.array((w,x,y,z))
        dist = np.linalg.norm(val_array-train_array)
        
        a_class = train.iloc[j]["Class"]
        output["value"] = dist
        output["name"] = a_class
        result.append(output)
    distance_iris.append(result)
distance_iris
```

```python
for i in range(len(validate)):
    distance_iris[i].sort(key=lambda k: k.get('value'))
for i in range(len(validate)):
    target = validate.iloc[i]["Class"]
    value_of_k =0
    for j in range(1,(len(train)+1)):
        k = j
        match , a = 0, 0
        wrong_match = 0
        for m in range(k):
            values = distance_iris[i][m]['name']
            if values == target:
                match += 1
            else:
                wrong_match += 1
        if(match>wrong_match):
             value_of_k =max(value_of_k,k)
print("Value Of K :- ",value_of_k)
```

```python
# Now match the predicted output with actual given output
test_distance = []
for i in range(len(test)):
    a = validate.iloc[i]["Sepal_Length"]
    b= validate.iloc[i]["Sepal_Width"]
    c = validate.iloc[i]["Petal_Length"]
    d = validate.iloc[i]["Petal_Width"]
    test_array = np.array((a,b,c,d))
    result = []
    for j in range(len(train)):
        output = {}
        w = train.iloc[j]["Sepal_Length"]
        x = train.iloc[j]["Sepal_Width"]
        y = train.iloc[j]["Petal_Length"]
        z = train.iloc[j]["Petal_Width"]
        train_array = np.array((w,x,y,z))
        dist = np.linalg.norm(test_array-train_array)
        
        a_class = train.iloc[j]["Class"]
        output["value"] = dist
        output["name"] = a_class
        result.append(output)
    test_distance.append(result)
for i in range(3):
    test_distance[i].sort(key=lambda k: k.get('value'))
```

```python
setosa_count = len(test[test["Class"]=="setosa"])
versicolor_count = len(test[test["Class"]=="versicolor"])
virginica_count = len(test[test["Class"]=="virginica"])

# calculate the count of predicte class for each type of class
test_setosa_count, test_versicolor_count, test_virginica_count = 0, 0, 0
for i in range(len(test_distance)):
    setosa , versicolor, virginica = 0, 0, 0
    for k in range(value_of_k):
        values = test_distance[i][k]['name']
        if(values == "setosa"):
            setosa+=1
        elif(values == "versicolor"):
            versicolor += 1
        else:
            virginica += 1
    if(setosa>=versicolor and setosa >= virginica):
        test_setosa_count+=1
    elif(versicolor>=setosa and versicolor>=virginica):
        test_versicolor_count+=1
    else:
        test_virginica_count += 1
if(setosa_count<= test_setosa_count):
    test_setosa_count = setosa_count
if(versicolor_count<=test_versicolor_count):
    test_versicolor_count = versicolor_count
if(virginica_count<=test_virginica_count):
    test_virginica_count = virginica_count
# print(setosa_count,versicolor_count,virginica_count)
# print(test_setosa_count, test_versicolor_count, test_virginica_count)
        
accuracy = (test_setosa_count + test_versicolor_count + test_virginica_count)/len(test)*100
print("Accuracy",accuracy,end="\n")
```
